```python
N = int(input())
A = list(map(int, input().split()))
colors = [0] * 8
all = 0
for a in A:
    if 1 <= a < 400:
        colors[0] += 1
    elif 400 <= a < 800:
        colors[1] += 1
    elif 800 <= a < 1200:
        colors[2] += 1
    elif 1200 <= a < 1600:
        colors[3] += 1
    elif 1600 <= a < 2000:
        colors[4] += 1
    elif 2000 <= a < 2400:
        colors[5] += 1
    elif 2400 <= a < 2800:
        colors[6] += 1
    elif 2800 <= a < 3200:
        colors[7] += 1
    else:
        all += 1

min_val = 0
for i in colors:
    if i != 0:
        min_val += 1

max_val = min_val + all
if min_val > 0:
    print(min_val, max_val)
else:
    print(1, all)
```
