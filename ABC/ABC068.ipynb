```python
from collections import deque
N, M = map(int, input().split())
G = [[] for _ in range(N + 1)]
edges = [list(map(int, input().split())) for i in range(M)]
for a, b in edges:
    G[a].append(b)
    G[b].append(a)

q = deque()
dist = [-1] * (N + 1)
dist[1] = 0
q.append(1)

while len(q) > 0:
    pos = q.popleft()
    for i in G[pos]:
        if dist[i] == -1:
            dist[i] = dist[pos] + 1
            q.append(i)

if dist[N] == 2:
    print('POSSIBLE')
else:
    print('IMPOSSIBLE')
    



```
