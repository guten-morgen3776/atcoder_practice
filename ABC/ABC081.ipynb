```python
N, K = map(int, input().split())
A = list(map(int, input().split()))
import collections
c = collections.Counter(A)
cnt = []
for i in c:
    cnt.append(c[i])

cnt = sorted(cnt, reverse=True)

ans = 0
while len(cnt) > K:
    ans += cnt.pop()
print(ans)


```
