```python
N = int(input())
from collections import deque
q = deque()
ans = True
t, x, y = map(int, input().split())
if t < (x + y) or t % 2 != (x + y) % 2:
    ans = False
q.append([t, x, y])
for _ in range(N - 1):
    t_, x_, y_ = q.popleft()
    t, x, y = map(int, input().split())
    T = t - t_
    d = abs(x - x_) + abs(y - y_)
    if T < d or T % 2 != d % 2:
        ans = False
        break
    q.append([t, x, y])
if ans == True:
    print('Yes')
else:
    print('No')
```
