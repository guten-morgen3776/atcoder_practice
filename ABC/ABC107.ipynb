```python
H, W = map(int, input().split())
grid = []
for _ in range(H):
    row = list(input())
    grid.append(row)
import numpy as np
grid = np.array(grid)

delete_row = []
for i in range(grid.shape[0]): #まず行チェック
    if '#' not in grid[i, :]:
        delete_row.append(i)
grid = np.delete(grid, delete_row, 0)

delete_column = []
for j in range(grid.shape[1]): #列チェック
    if '#' not in grid[:, j]:
        delete_column.append(j)
grid = np.delete(grid, delete_column, 1)

for row in grid:
    print(''.join(row))

```
