```python
N = int(input())
P = list(map(int, input().split()))
Q = list(map(int, input().split()))
P = tuple(P)
Q = tuple(Q)
from itertools import permutations
P_ = sorted(list(permutations(P)))
Q_ = sorted(list(permutations(Q)))
a = 0
for i in range(len(P_)):
    if P_[i] == P:

        a = i + 1
b = 0
for j in range(len(Q_)):
    if Q_[j] == Q:
        b = j + 1
print(abs(b - a))
```
