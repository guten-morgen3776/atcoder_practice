```python
N, M = map(int, input().split())
ans = [-1] * N
possible = True
for _ in range(M):
    s, c = map(int, input().split())
    if ans[s - 1] != -1 and ans[s - 1] != c:
        possible = False
        break
    ans[s - 1] = c

if possible:
    if N > 1 and ans[0] == 0:
        print(-1)
    else:
        if ans[0] == -1:
            if N == 1:
                ans[0] = 0
            else:
                ans[0] = 1
        for i in range(1, N):
            if ans[i] == -1:
                ans[i] = 0
        print(''.join(map(str, ans)))
else:
    print(-1)
```
