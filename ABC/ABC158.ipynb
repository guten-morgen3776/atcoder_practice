```python
from collections import deque

S = input()
Q = int(input())
S = deque(S)
rev = 1 #1なら普通、−１なら反転
for _ in range(Q):
    que = list(input().split())
    if que[0] == '1':
        rev *= -1
    else:
        if que[1] == '1' and rev == 1:
            S.appendleft(que[2])
        elif que[1] == '1' and rev == -1:
            S.append(que[2])
        elif que[1] == '2' and rev == 1:
            S.append(que[2])
        else:
            S.appendleft(que[2]) 

if rev == 1:
    print(''.join(S))
else:
    S = list(S)
    S = S[::-1]
    print(''.join(S))
```
