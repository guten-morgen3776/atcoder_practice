```python
N = int(input())
A = list(map(int, input().split()))
from collections import Counter
c = Counter(A)
A_count = [c[i] for i in A]
all = 0
for val in c:
    n = c[val]
    all += n * (n - 1) // 2

for i in range(N):
    a = A_count[i]
    if a == 1:
        A_count[i] = 0
    else:
        A_count[i] = a * (a - 1) // 2 - (a - 1) * (a - 2) // 2

for i in range(N):
    ans = all - A_count[i]
    print(ans)
```
