```python
N, M, X = map(int, input().split())
A = []
for _ in range(N):
    row = list(map(int, input().split()))
    A.append(row)

min_cost = float('inf')
for i in range(1 << N):
    cur_subset = []
    for j in range(N): #N桁のビットをチェック
        if (i & (1 << j)):
            cur_subset.append(j)

    skills = [0] * M
    for l in range(1, M + 1):
        for m in cur_subset:
            skills[l - 1] += A[m][l]

    if min(skills) >= X:
        cost = 0
        for k in cur_subset:
            cost += A[k][0]
        min_cost = min(min_cost, cost)
if min_cost == float('inf'):
    print(-1)
else:
    print(min_cost)


```
