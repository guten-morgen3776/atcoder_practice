```python
N = int(input())
S = list(map(int, input().split()))
T = list(map(int, input().split()))
first_times = []
first = T[0]
first_times.append(first)

for i in range(N - 1):
    first = min(first + S[i], T[i + 1])
    first_times.append(first)

first = min(first_times[0], first_times[N - 1] + S[N - 1])

first_times[0] = first
for j in range(N - 1):
    first = min(first + S[j], T[j + 1])
    first_times[j + 1] = min(first, first_times[j + 1])

for k in range(N):
    print(first_times[k])


```
