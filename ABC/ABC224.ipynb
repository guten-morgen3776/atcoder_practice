```python
N = int(input())
grids = []
for _ in range(N):
    x, y = map(int, input().split())
    grids.append([x, y])
import itertools
selected = list(itertools.combinations(grids, 3))
ans = 0
for i in range(len(selected)):
    three = selected[i]
    p1 = three[0]
    p2 = three[1]
    p3 = three[2]
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    if (x2 - x1) * (y3 - y1) == (x3 - x1) * (y2 - y1):
        continue
    ans += 1
print(ans)
```
