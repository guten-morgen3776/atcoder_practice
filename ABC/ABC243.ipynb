```python
N, Q = map(int, input().split())
parts = [[i + 1, 0] for i in range(N)]

from collections import deque
parts = deque(parts)
for _ in range(Q):
    que = list(input().split())
    if que[0] == '1':
        x, y = parts[0]
        if que[1] == 'R':
            x += 1
        elif que[1] == 'L':
            x -= 1
        elif que[1] == 'U':
            y += 1
        else:
            y -= 1
        parts.appendleft([x, y])
    else:
        print(*parts[int(que[1]) - 1])

```
