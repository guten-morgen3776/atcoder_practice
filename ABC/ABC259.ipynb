```python
S = list(input())
T = list(input()) 
from itertools import groupby
RLS = [(k, len(list(g))) for k, g in groupby(S)]
RLT = [(k, len(list(g))) for k, g in groupby(T)]  

if len(RLS) != len(RLT):
    print('No')
else:
    for i in range(len(RLS)):
        char_s, count_s = RLS[i]
        char_t, count_t = RLT[i]
        if char_s != char_t:
            print('No')
            break
        if count_s > count_t:
            print('No')
            break
        if count_s == 1 and count_t > 1:
            print('No')
            break
    else:
        print('Yes')




```
