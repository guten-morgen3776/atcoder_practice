```python
N, X = map(int, input().split())
coins = []
for _ in range(N):
    ab = list(map(int, input().split()))
    coins.append(ab)

coin_sort = sorted(coins, reverse=True, key=lambda x: x[0])
import numpy as np
coin_sort = np.array(coin_sort)
rem = X
ans = -1
while rem > 0:
    for i in range(N):
        if rem >= coin_sort[i, 0] and coin_sort[i, 1] > 0:
            rem -= coin_sort[i, 0]
            coin_sort[i, 1] -= 1
            continue
    if rem == 0:
        ans = 1
    elif 0 < rem < min(coin_sort[:, 0]):
        ans = 0

if ans == 1:
    print('Yes')
else:
    print('No')
```
