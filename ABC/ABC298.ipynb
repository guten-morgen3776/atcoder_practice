```python
N = int(input())
Q = int(input())
boxes = [[] for i in range(N)]
numbers = [[] for i in range(2 * (10 ** 5))]

for _ in range(Q):
    que = list(map(int, input().split()))
    i = que[1]
    if len(que) == 3:
        j = que[2]

    if que[0] == 1:
        boxes[j - 1].append(i)
        if j not in numbers[i - 1]:
            numbers[i - 1].append(j)
    elif que[0] == 2:
        box_sort = sorted(boxes[i - 1])  
        print(*box_sort)
    else:
        number_sort = sorted(numbers[i - 1])
        print(*number_sort)

```
