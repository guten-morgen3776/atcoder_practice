```python
N, M, H, K = map(int, input().split())
S = list(input())
from collections import defaultdict
grid = defaultdict(lambda: defaultdict(int))
for _ in range(M):
    x, y = map(int, input().split())
    grid[x][y] = K

cur = [0, 0]
vit = H
for i in range(N):
    if S[i] == 'R':
        cur[0] += 1
        vit -= 1
        if vit < 0:
            break
        if grid[cur[0]][cur[1]] == K and vit < K:
            vit = K
            grid[cur[0]][cur[1]] = 0
        
    elif S[i] == 'L':
        cur[0] -= 1
        vit -= 1
        if vit < 0:
            break
        if grid[cur[0]][cur[1]] == K and vit < K:
            vit = K
            grid[cur[0]][cur[1]] = 0
        
    elif S[i] == 'U':
        cur[1] += 1
        vit -= 1
        if vit < 0:
            break
        if grid[cur[0]][cur[1]] == K and vit < K:
            vit = K
            grid[cur[0]][cur[1]] = 0
        
    elif S[i] == 'D':
        cur[1] -= 1 
        vit -= 1
        if vit < 0:
            break
        if grid[cur[0]][cur[1]] == K and vit < K:
            vit = K
            grid[cur[0]][cur[1]] = 0
        
  

if vit < 0:
    print('No')
else:
    print('Yes')
```
