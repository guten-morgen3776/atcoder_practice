```python
import sys
import functools
sys.setrecursionlimit(200010)

def compare(item_i, item_j):
    Ai, Bi, index_i = item_i
    Aj, Bj, index_j = item_j
    val_i = Ai * Bj
    val_j = Aj * Bi
    if val_i > val_j:
        return -1
    elif val_i < val_j:
        return 1
    else:
        return 0

N = int(input())
people = []
for i in range(N):
    a, b = map(int, input().split()) 
    people.append((a, b, i + 1))

key_function = functools.cmp_to_key(compare)
sorted_people = sorted(people, key=key_function)
result_indices = [item[2] for item in sorted_people]
print(*result_indices)


```

```python
H, W = map(int, input().split())
grid = []
for _ in range(H):
    row = list(input())
    grid.append(row)
from collections import deque
dist = []
for _ in range(H):
    row = [-1] * W
    dist.append(row)
q = deque()
q.append([0, 0])
dist[0][0] = 0
if grid[0][0] == 's':
    while len(q) > 0:
        pos = q.popleft()
        y = pos[0]
        x = pos[1]
        nexts = [[y+1, x], [y, x+1], [y-1, x], [y, x-1]]
        for next in nexts:
            yi = next[0]
            xi = next[1]
            if 0 <= yi <= H - 1 and 0 <= xi <= W - 1 and dist[yi][xi] == -1:
                mod = (dist[y][x] + 1) % 5
                if mod == 0 and grid[yi][xi] == 's':
                    dist[yi][xi] = dist[y][x] + 1
                    q.append(next)
                elif mod == 1 and grid[yi][xi] == 'n':
                    dist[yi][xi] = dist[y][x] + 1
                    q.append(next)
                elif mod == 2 and grid[yi][xi] == 'u':
                    dist[yi][xi] = dist[y][x] + 1
                    q.append(next)
                elif mod == 3 and grid[yi][xi] == 'k':
                    dist[yi][xi] = dist[y][x] + 1
                    q.append(next)
                elif mod == 4 and grid[yi][xi] == 'e':
                    dist[yi][xi] = dist[y][x] + 1
                    q.append(next)          

if dist[H-1][W-1] != -1:
    print('Yes')
else:
    print('No')
```
