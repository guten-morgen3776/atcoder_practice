```python
N1, N2, M = map(int, input().split())
N = N1 + N2
edges = [list(map(int, input().split())) for i in range(M)]
G = [list() for i in range(N + 1)]
for a, b in edges:
    G[a].append(b)
    G[b].append(a)
from collections import deque
#1からの最短距離が最大、Nからの距離が最大をそれぞれ求めて繋げばいい
q1 = deque()
dist1 = [-1] * (N + 1)
dist1[1] = 0
q1.append(1)

while len(q1) > 0:
    pos = q1.popleft()
    for i in G[pos]:
        if dist1[i] == -1:
            dist1[i] = dist1[pos] + 1
            q1.append(i)
d1 = max(dist1)

q2 = deque()
dist2 = [-1] * (N + 1)
dist2[N] = 0
q2.append(N)

while len(q2) > 0:
    pos = q2.popleft()
    for i in G[pos]:
        if dist2[i] == -1:
            dist2[i] = dist2[pos] + 1
            q2.append(i)
d2 = max(dist2)

print(d1 + d2 + 1)

```
