```python
N, Q = map(int, input().split())
R = list(map(int, input().split()))
R_sort = sorted(R)

rui = [0] * (N + 1)
for i in range(1, N + 1):
    rui[i] = rui[i - 1] + R_sort[i - 1]

import bisect
for _ in range(Q):
    x = int(input())
    ans = bisect.bisect_right(rui, x) #right,leftは含む含まないで使い分け
    print(ans - 1)
    
```
