```python
a, b, c, d, e, f = map(int, input().split())
g, h, i, j, k, l = map(int, input().split())
cnt_x = 0
for x in range(a + 1, d):
    if x in range(g, j + 1):
        cnt_x += 1
cnt_y = 0
for y in range(b + 1, e):
    if y in range(h, k + 1):
        cnt_y += 1
cnt_z = 0
for z in range(c + 1, f):
    if z in range(i, l + 1):
        cnt_z += 1
if cnt_x > 0 and cnt_y > 0 and cnt_z > 0:
    print('Yes')
else:
    print('No')
```
