```python
N = int(input())
H = list(map(int, input().split()))

ans = 1
for k in range(1, N):
    for start in range(k):
        height = 0
        current = 0
        for i in range(start, N, k): #k飛ばしで進んでいく
            if height != H[i]:
                current = 0
                height = H[i]
            current += 1
            ans = max(ans, current)

print(ans)
```
