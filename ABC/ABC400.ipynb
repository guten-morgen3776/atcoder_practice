```python
"""
素因数分解的な感じ
N<=10**18はめっちゃでかいな
毎回計算してたらTLEだし
なんなら事前に計算していても全てのNについて検証するのは不可能
これaの範囲は結構狭いのでは？ 多分50後半まで
aが確定すればN // 2**a以下で最大のb**2を見つければ個数がわかる
〜以下で最大だからリストにして二分探索でいけるはず
O(logN * logN)より小さくなるはず
そもそもルートとって少数切り捨てればいいのでは？
これ正数aじゃないといけないのか 
これ普通に数えたらめっちゃ被るやん
でもこれ毎回setに入れて管理してたら間に合わないよな
これどんな場合でもbをでかくすればaは1,2に帰着できるやん
"""
"""
浮動小数点によるずれを回避するためにmath.isqrt()を使う
"""

import math
N = int(input())
ans = 0
for a in range(1, 3):
    p = N // (2 ** a)
    if p > 0:
        max_b = math.isqrt(p)
        ans += max_b
print(ans)
```
