```python
#前K項の和が次のこう
#K項分のわを保持し続ければいい
#移動のたびに差分だけを考える
#差分を考えるならdequeとしてリストを保持した方がいい
#毎回sumはTLE 予め計算して差分を考える

from collections import deque
q = deque()
mod = 10 ** 9
N, K = map(int, input().split())
A = [1] * (N + 1)
for i in range(K):
    q.append(1)
sum_A = K
for i in range(K, N + 1): 
    new = sum_A 
    sum_A += (new - q[0]) % mod
    q.append(new)
    q.popleft()
if N >= K:
    print(new % mod)
else:
    print(1)
```
