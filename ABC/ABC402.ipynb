```python
N, M = map(int, input().split())
gyakubiki = [[] for _ in range(N + 1)] #日から食材ではなく食材から日
nigate = [0] * M #苦手な食材の数（料理ごと）
for i in range(M):
    row = list(map(int, input().split()))
    k = row[0]
    syokuzai = row[1:]
    nigate[i] = k
    for j in syokuzai: 
        gyakubiki[j].append(i)
#print(gyakubiki) 

B = list(map(int, input().split()))
ans = 0

for i in range(N):
    b = B[i]
    for j in gyakubiki[b]:
        nigate[j] -= 1
        if nigate[j] == 0:
            ans += 1
    print(ans)


      
    
```
