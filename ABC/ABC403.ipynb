```python
N, M, Q = map(int, input().split())
from collections import defaultdict
can_see = defaultdict(lambda: defaultdict(lambda: 0))
see_all = [0] * (N + 1)
for _ in range(Q):
    que = list(map(int, input().split()))
    if que[0] == 1:
        x, y = que[1], que[2]
        can_see[x][y] = 1
    elif que[0] == 2:
        x = que[1]
        see_all[x] = 1
    else:
        x, y = que[1], que[2]
        if see_all[x] == 1:
            print('Yes')
        elif can_see[x][y] == 1:
            print('Yes')
        else:
            print('No')

```
