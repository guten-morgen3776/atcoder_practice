```python

class UnionFind:
	def __init__(self, n):
		#親要素のノード番号を格納　根は−１
		self.parents = [-1] * n
		
	def find(self, x): #各番号の根がどれかを探す
		if self.parents[x] < 0:
			return x
		else:
			#再起処理　親の親とたどって根を見つける
			self.parents[x] = self.find(self.parents[x])
			return self.parents[x]
			
	def union(self, x, y):
		#お互いの根を呼び出す
		x = self.find(x)
		y = self.find(y)
		if x == y:
			return False
		if self.parents[x] > self.parents[y]:
		#parentsの値はマイナス　絶対値が人数
			x, y = y, x
		self.parents[x] += self.parents[y]	
		self.parents[y] = x
		return True
	
	def is_same(self, x, y):
		return self.find(x) == self.find(y)	
	
    

#判定
def is_cycle_graph(n, m, edges):
	if n != m:
		return False
	

N, M = map(int, input().split())
edges = []
for _ in range(M):
	u, v = map(int, input().split())
	edges.append((u-1, v-1))
if is_cycle_graph(N, edges) == True:
	print('Yes')
else:
	print('No')
```

```python
N, M = map(int, input().split())
edges = [list(map(int, input().split())) for i in range(M)]

G = [list() for i in range(N + 1)]
for a, b in edges:
    G[a].append(b)
    G[b].append(a)

import sys
sys.setrecursionlimit(120000)

visited = [False] * (N + 1)
if N != M:
    print('No')
    sys.exit()

for i in range(1, N + 1):
    if len(G[i]) != 2:
        print('No')
        sys.exit()


def dfs_count(G, pos, visited):
    visited[pos] = True
    cnt = 1
    for i in G[pos]:
        if visited[i] == False:
            cnt += dfs_count(G, i, visited)
    return cnt

if dfs_count(G, 1, visited) == N:
    print('Yes')
else:
    print('No')
```
