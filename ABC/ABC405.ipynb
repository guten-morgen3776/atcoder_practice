```python
N = int(input())
A = list(map(int, input().split()))

rui = [0] * (N + 1)
for i in range(1, N + 1):
    rui[i] = rui[i - 1] + A[i - 1]

ans = 0
for i in range(N):
    ans += A[i] * (rui[N] - (rui[N] - rui[i]))
print(ans)
```

```python
H, W = map(int, input().split())
grid = [list(input()) for i in range(H)]
dist = [[-1] * W for i in range(H)]

from collections import deque

ans = grid.copy()

for i in range(H):
    for j in range(W):
        if grid[i][j] == '.':
            q = deque()
            q.append([i, j])
            while len(q) > 0:
                pos = q.popleft()
                y, x = pos[0], pos[1]
                next = [[y + 1, x],[y, x + 1], [y - 1, x],[y, x - 1]]
                for k in next:
                    yi, xi = k[0], k[1]
                    if dist[yi][xi] == -1 and grid[yi][xi] == '.':
                        dist[yi, xi] = dist[y, x] + 1
                        q.append(k)

```
