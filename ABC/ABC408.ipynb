```python
N, M = map(int, input().split())
houdai = [0] * N 
diff = [0] * (N + 1)
queries = []
for _ in range(M):
    que = list(map(int, input().split()))
    queries.append(que)

for l, r in queries:
    diff[l - 1] += 1
    diff[r] -= 1

cur_val = 0
for i in range(N):
    cur_val += diff[i]
    houdai[i] = cur_val
print(min(houdai))
```
