```python
N, L  = map(int, input().split())
d = list(map(int, input().split()))

if L % 3 != 0:
    print(0)
else:
    place = [0] * L
    place[0] += 1
    point = 0
    for i in range(N - 1):
        point += d[i]
        point %= L
        place[point] += 1
    
    ans = 0
    for k in range(L // 3):
        ans += place[k] * place[k + L // 3] * place[k + 2 * L // 3]
    print(ans)
```
