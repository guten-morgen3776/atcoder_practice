```python
N, Q  = map(int, input().split())
A = [i for i in range(1, N + 1)]

zure_idx = 0
for _ in range(Q):
    que = list(map(int, input().split()))
    if que[0] == 1:
        p = que[1]
        x = que[2]
        real_idx = (zure_idx + (p - 1)) % N
        A[real_idx] = x
    elif que[0] == 2:
        p = que[1]
        real_idx = (zure_idx + (p - 1)) % N
        print(A[real_idx])
    else:
        k = que[1]
        zure_idx = (zure_idx + k) % N




```
