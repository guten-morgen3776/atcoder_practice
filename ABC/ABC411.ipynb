```python
N = int(input())
D = list(map(int, input().split()))
RUI = [0] * N
for i in range(1, N):
    rui = RUI[i - 1] + D[i - 1]
    RUI[i] = rui

for i in range(0, N - 1):
    distance = []
    for j in range(1, N - i):
        dis = RUI[i + j] - RUI[i]
        distance.append(dis)
    print(*distance)
```

```python
N, Q = map(int, input().split())
A = list(map(int, input().split()))

from collections import defaultdict
masu = defaultdict(lambda: 1)

from collections import deque
q = deque()
q.append(masu)
#siro:1, kuro:-1
ans = 0
for i in range(Q):
    masu_prev = q.popleft()
    if masu[A[i] - 1] == 1 and masu_prev[A[i] - 2] == 1 and masu_prev[A[i]] == 1:
        ans += 1
    elif masu[A[i] - 1] == -1 and masu_prev[A[i] - 2] == -1 and masu_prev[A[i]] == -1:
        ans += 1
    elif masu[A[i] - 1] == -1  and masu_prev[A[i] - 2] == 1 and masu_prev[A[i]] == 1:
        ans -= 1
    elif masu[A[i] - 1] == 1 and masu_prev[A[i] - 2] == -1 and masu_prev[A[i]] == -1:
        ans -= 1
    masu[A[i] - 1] *= -1
    q.append(masu)
    print(ans)
```
