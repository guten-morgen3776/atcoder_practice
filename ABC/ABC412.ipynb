```python
T = int(input())
for _ in range(T):
    N = int(input())
    S = list(map(int, input().split()))
    if N == 1:
        print(1)
    else:
        cur_idx = 0
        cnt = 1
        while cur_idx < N - 1:
            Si = S[cur_idx]
            if 2 * Si >= S[N - 1]:
                cnt += 1
                break
            next_j = -1
            for j in range(cur_idx + 1, N - 1):
                if 2 * Si >= S[j]:
                    next_j = j
            if next_j == -1:
                cnt = -1
                break
            if Si >= S[next_j]:
                cnt = -1
                break
            cur_idx = next_j
            cnt += 1
        print(cnt)


```

```python
"""
大きさが2倍以下なら倒れる
できるだけ少ない数で最後まで倒れるようにしたい
次のドミノをできるだけ大きくしたいから二倍以下で一番大きいものを探す
でも各ケースについてやってたら間に合わなくね？
でもNのテストケースでの和が定まってるから全てについてみても間に合うか
これスタートとゴールはソートする前のものじゃん
スタートとゴールだけ取ってそれ以外で間をソートして現在の二倍以下で最大ものを2分探索で求める
O(NlogN)だからいける
"""
import bisect
T = int(input())
for _ in range(T):
    N = int(input())
    S = list(map(int, input().split()))
    start = S[0]
    goal = S[N - 1]
    S_new = []
    S_new.append(S[0])
    if goal > start:
        for i in range(N):
            if goal > S[i] > start:
                S_new.append(S[i])
    else:
        print(2)
        continue
    S_new.append(goal)
    S_new.sort()
    cur_idx = 0
    ans = 1
    possible = True
    while S_new[cur_idx] < goal:
        max_idx = bisect.bisect_right(S_new, 2 * S_new[cur_idx]) - 1
        if max_idx == cur_idx:
            possible = False
            break
        cur_idx = max_idx
        ans += 1
    if possible:
        print(ans)
    else:
        print(-1)

```
