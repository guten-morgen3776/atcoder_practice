```python
N = int(input())
S = []
for _ in range(N):
    s = input()
    S.append(s)
count = 0
CONC = []
for i in range(N):
    for j in range(N):
        if i == j:
            continue
        conc = S[i] + S[j] 
        if conc in CONC:
            count = count 
        else:
            CONC.append(conc)
            count += 1
print(count)          
```
