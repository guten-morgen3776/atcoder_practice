```python
N = int(input())
C = []
L = []
for _ in range(N):
    c, l = input().split()
    C.append(c)
    L.append(int(l))

count_len = 0
out = 0
S = []
for i in range(N):
    count_len += L[i]
    if count_len > 100:
        out += 1
        print('Too Long')
        break
    else:
        S.append(C[i] * L[i])
        continue

if out == 0:
    print(*S, sep='')
```
