```python
N, K, X = map(int, input().split())
strings = []
for i in range(N):
    string = list(input())
    strings.append(string)
from itertools import product
conc_str = list(product(strings, repeat=K))

joined = []
for j in range(len(conc_str)):
    temp_str = []
    for k in range(K):
        temp_str += conc_str[j][k] 
    joined.append(temp_str)
joined.sort()
print(''.join(joined[X - 1]))
```
