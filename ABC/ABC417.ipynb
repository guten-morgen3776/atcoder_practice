```python
N = int(input())
A = list(map(int, input().split()))
j_Aj = []
for j in range(1, N + 1):
    j_Aj.append(j - A[j - 1])
i_Ai = []
for i in range(1, N + 1):
    i_Ai.append(i + A[i - 1])

from collections import defaultdict
count_left = defaultdict(int)
for i in range(1, N + 1):
    vali = i_Ai[i - 1]
    count_left[vali] += 1
ans = 0
for j in range(1, N + 1):
    valj = j_Aj[j - 1]
    ans += count_left[valj]
print(ans)
```
