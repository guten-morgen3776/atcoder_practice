```python
#ディーラーはできるだけ異なるフレーバーを渡したい
#同じものをb個以上入れざる終えなくなれば勝ち
#Aを縦に並べて下から横方向にとっていけば良い　
#Aを横向きに0,1の行列としてとって転置して和を取り直す
#でも行列に直す際にO(AN)だからTLE
#min(Ai, b)の和を取れば実質横からとっているもの
#具体的にはmin(Ai, b-1)の和＋１
#でも普通に和をとったらO(NQ)でTLE
#境界を二分探索で求めて累積和 O(QlogN)


N, Q = map(int, input().split())
A = list(map(int, input().split()))
A.sort()
rui = [0] * (N + 1)
for i in range(1, N + 1):
    rui[i] = A[i - 1] + (rui[i - 1] if i > 1 else 0) 

import bisect
for _ in range(Q):
    b = int(input())
    border = bisect.bisect_right(A, b - 1) 
    ans = rui[border] + (b - 1) * (N - border) + 1
    if b > A[N - 1]:
        print(-1)
    else:
        print(ans)
```
