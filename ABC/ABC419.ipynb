```python
#行き先が決まれば移動回数はもとまる
#一番離れている人が分かればいい
#一番離れているペアの中心をゴールにすればいいのでは？
#全員の距離を測っているとO(N^2)になってしまう
#そもそもこのグリッドでの距離とはmax(abs(Ri - R), abs(Ci - C))
#だからR,Cを独立に考えて良い
#最大最小のR, Cを保持していればどれで距離の最大がわかる
N = int(input())
R = []
C = []
for _ in range(N):
    r, c = map(int, input().split())
    R.append(r)
    C.append(c)
max_r = max(R)
min_r = min(R)
max_c = max(C)
min_c = min(C)

dist = max(abs(max_r - min_r), abs(max_c - min_c))
print((dist + 1) // 2)
```

```python
"""
入れ替える回数がM回
少なくともO(M)は必要 
一つ同士ならO(1)で入れ替えできるので範囲ありで一括入れ替えできればそれでできそうだが
リストとして管理すべき？
リストならな簡単に範囲指定で入れ替えできるわ
入れ替え操作はO(k)らしい つまり毎回入れ替えていたら無理なわけだ
入れ替えた回数が奇数回ならT、偶数回ならSになるから入れ替え回数を記録できればいい
範囲内に一括で＋１すればいい これは芋酢でできたはず
"""
N, M = map(int, input().split())
S = list(input())
T = list(input())
swap = [0] * (N + 1)
for _ in range(M):
    l, r = map(int, input().split())
    swap[l - 1] += 1
    swap[r] -= 1

for i in range(1, N + 1):
    swap[i] += swap[i-1]

ans = []
for j in range(N):
    if swap[j] % 2 == 0:
        ans.append(S[j])
    else:
        ans.append(T[j])
print(''.join(map(str, ans)))
```
