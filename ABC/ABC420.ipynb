```python
N, Q = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
ans = 0
for i in range(N):
    ans += min(A[i], B[i])

for _ in range(Q):
    c, X, V = input().split()
    X = int(X)
    V = int(V)
    m = min(A[X-1], B[X-1])
    ans -= m
    if c == 'A':
        A[X-1] = V
    else:
        B[X-1] = V
    ans += min(A[X-1], B[X-1])
    print(ans)
```
