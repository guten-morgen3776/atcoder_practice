```python
N = int(input())
from collections import deque
q = deque()
visited = [False] * (N + 1)

graph = [[] for i in range(N + 1)]
for i in range(1, N + 1):
    a, b = map(int, input().split())
    if a == 0 and b == 0:
        q.append(i)
        visited[i] = True
    else:
        if a != 0:
            graph[a].append(i)
        if b != 0:
            graph[b].append(i)

while len(q) > 0:
    pos = q.popleft()
    for i in graph[pos]:
        if  visited[i] == False:
            visited[i] = True
            q.append(i)

print(sum(visited))

```
