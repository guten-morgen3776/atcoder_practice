```python
N, M = map(int, input().split())
edges = []
for _ in range(M):
    u, v = map(int, input().split())
    edges.append((u-1, v-1))

#bitはN個の頂点の塗り分けかた 
"""
Nこの頂点を2色で塗り分けていく そして同じ色で塗られた頂点同士を結ぶ辺の数を調べて
それの最小値を求める
bitは10進法で頂点の塗り分け方を表している 

<<,>>はビットシフト
1 << N は２**N
(bit >> u) & 1は頂点uの色は何か(0,1)を調べる
"""   
ans = M
for bit in range(1 << N): 
    delete_count = 0
    for u, v in edges:
        color_u = (bit >> u) & 1
        color_v = (bit >> v) & 1
        if color_u == color_v:
            delete_count += 1
    ans = min(ans, delete_count)
print(ans)
```
