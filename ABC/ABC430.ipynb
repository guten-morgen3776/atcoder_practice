```python
N, A, B = map(int, input().split())
S = list(input())
print(S)

right = 0
cnt_a = 0
cnt_b = 0
ans = 0

for left in range(N): #左を動かす
    while right < N: #右を動かす
        if S[right] == 'b' and cnt_b + 1 >= B:
            break
        if S[right] == 'a':
            cnt_a += 1
        elif S[right] == 'b':
            cnt_b += 1
        right += 1
    print(f'left: {left}, right: {right - 1}, cnt_a: {cnt_a}, cnt_b: {cnt_b}')
    if cnt_a >= A:
        ans += 1
        print('OK')
    if left == right:
        right += 1
    else:
        if S[left] == 'a':
            cnt_a -= 1
        elif S[left] == 'b':
            cnt_b -= 1
        

print(ans)
```

```python
"""
区間ごとのa,bの数が分かればいい
それは累積和でいい
lを固定してaがA以上になる最小,bがB未満になる最大を求めてその間がOK
O(NlogN)だからOK
"""
N, A, B = map(int, input().split())
S = list(input())
cnt_a = [0] * (N + 1)
cnt_b = [0] * (N + 1)

if S[0] == 'a':
    cnt_a[1] = 1
    cnt_b[1] = 0
else:
    cnt_a[1] = 0
    cnt_b[1] = 1

for i in range(1, N):
    if S[i] == 'a':
        cnt_a[i] = cnt_a[i - 1] + 1
        cnt_b[i] = cnt_b[i - 1]
    else:
        cnt_a[i] = cnt_a[i - 1] 
        cnt_b[i] = cnt_b[i - 1] + 1

import bisect
ans = 0
for l in range(1, N + 1):
    min_r = bisect.bisect_left(cnt_a, A + cnt_a[l - 1]) - 1
    max_r = bisect.bisect_right(cnt_b, B + cnt_b[l - 1]) - 1
    print(l, min_r, max_r)
    if max_r >= min_r:
        ans += max_r - min_r + 1
print(ans)

        
```
