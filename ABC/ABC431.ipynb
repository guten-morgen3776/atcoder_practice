```python
X = int(input())
N = int(input())
W = list(map(int, input().split()))
Q = int(input())
weight = X
parts = []
for _ in range(Q):
    p = int(input())
    if p not in parts:
        parts.append(p)
        weight += W[p - 1]
    else:
        weight -= W[p - 1]
        parts.remove(p)
    print(weight)
```

```python
N, M, K = map(int, input().split())
H = list(map(int, input().split()))
B = list(map(int, input().split()))
H_sort = sorted(H)
B_sort = sorted(B)

robot = 0
i = 0
j = 0
while i < N and j < M:
    if H_sort[i] <= B_sort[j]:
        robot += 1
        i += 1
        j += 1
    else:
        j += 1
if robot >= K:
    print('Yes')
else:
    print('No')



```

```python
N = int(input())
w_total = 0
D = []
W = []
Sb = 0
for _ in range(N):
    w, h, b = map(int, input().split()) 
    D.append(h - b)
    w_total += w
    W.append(w)
    Sb += b
dp = []
for _ in range(N + 1):
    row = [-float('inf')] * (w_total + 1)
    dp.append(row)
dp[0][0] = 0
for i in range(1, N + 1):
    wi = W[i - 1]
    di = D[i - 1]
    for w in range(w_total // 2 + 1):
        not_use = dp[i - 1][w]
        use = -float('inf')
        if w - wi >= 0:
            use = dp[i - 1][w - wi] + di
        dp[i][w] = max(not_use, use)

max_D = -float('inf')
for w in range(w_total // 2 + 1):
    max_D = max(max_D, dp[N][w])
print(Sb + max_D)



        
```
