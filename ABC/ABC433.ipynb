```python
S = list(map(int, input()))
diff_1 = []
for i in range(len(S) - 1):
    if S[i] + 1 == S[i + 1]:
        diff_1.append(i)

ans = 0
if len(diff_1) == 0:
    ans = 0
else:
    ans = len(diff_1)
    for i in diff_1:
        left = i
        right = i + 1
        while left - 1 >= 0 and right + 1 <= len(S) - 1 and S[left] == S[left - 1] and S[right] == S[right + 1]:
            ans += 1
            left -= 1
            right += 1

print(ans) 
```

```python
N, M = map(int, input().split())
A = list(map(int, input().split()))

from collections import defaultdict
count = [defaultdict(int) for _ in range(10)]
for i in range(N):
    mod = A[i] % M
    keta = len(str(A[i]))
    count[keta - 1][mod] += 1

ans = 0
for x in A:
    for k in range(1, 11):
        mod_A = x * 10 ** k % M
        ans += count[k - 1][(M - mod_A) % M]

print(ans)
```
