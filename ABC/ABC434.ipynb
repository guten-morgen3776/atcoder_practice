```python
#ti->ti+1の間に範囲内にいられればいい
#現在位置を保持している必要ある？
#現在位置+-時間差が次の範囲に含まれていればいい
#でも範囲でやってると現在位置を特定できない
#でも次のことだけ考えて移動すればいいわけじゃなくて
#２手先がギリギリならできるだけ事前に移動しておかないといけない
#到達可能範囲と目標範囲の共通部分が存在範囲

T = int(input())
for _ in range(T):
    N, H = map(int, input().split())
    pre_t = 0
    pre_l = H
    pre_u = H
    ans = True
    for _ in range(N):
        t, l, u = map(int, input().split())
        dt = t - pre_t
        cur_l = max(pre_l - dt, l)
        cur_u = min(pre_u + dt, u)
        if cur_l > cur_u:
            ans = False
        pre_t = t
        pre_l = cur_l
        pre_u = cur_u
    if ans == True:
        print('Yes')
    else:
        print('No')
```

```python
N = int(input())
grid = [[0] * (2002) for _ in range(2002)]
clouds = []
for _ in range(N):
    u, d, l, r = list(map(int, input().split()))
    clouds.append([u, d, l, r])
    grid[u][l] += 1
    grid[u][r + 1] -= 1
    grid[d + 1][l] -= 1
    grid[d + 1][r + 1] += 1

#横方向に累積和
for i in range(1, 2002):
    for j in range(1, 2002):
        grid[i][j] += grid[i][j - 1]

#縦方向に累積わ
for j in range(1, 2002):
    for i in range(1, 2002):
        grid[i][j] += grid[i - 1][j]

empty_count = 0
ones_grid = [[0] * (2002) for _ in range(2002)]
for i in range(1, 2001):
    for j in range(1, 2001):
        if grid[i][j] == 0:
            empty_count += 1
        elif grid[i][j] == 1:
            ones_grid[i][j] = 1

S = [[0] * (2002) for _ in range(2002)]
for i in range(1, 2001):
    for j in range(1, 2001):
        S[i][j] = S[i - 1][j] + S[i][j - 1] - S[i - 1][j - 1] + ones_grid[i][j]

for u, d, l, r in clouds:
    count_1 = S[d][r] - S[u - 1][r] - S[d][l - 1] + S[u - 1][l - 1]
    print(empty_count + count_1)



```
