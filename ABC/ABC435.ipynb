```python
N = int(input())
A = list(map(int, input().split()))
rui = [0] * N
rui[0] = A[0]
for k in range(1, N):
    rui[k] = rui[k - 1] + A[k]

ans = 0
for l in range(1, N + 1):
    for r in range(l, N + 1):
        all = 0
        for i in range(l, r + 1):
            if l >= 2:
                sum_A = rui[r - 1] - rui[l - 2]
                if sum_A % A[i - 1] != 0:
                    all += 1
            elif l == 1:
                sum_A = rui[r - 1]
                if sum_A % A[i - 1] != 0:
                    all += 1
        if all == r - l + 1:
            ans += 1
    
                
print(ans)
```

```python
N = int(input())
A = list(map(int, input().split()))

cnt = A[0]
biggest = A[0]
if A[0] != 1:
    for i in range(1, N):
        biggest -= 1
        if biggest <= A[i]:
            cnt += A[i] - biggest
            biggest = A[i]
        if biggest == 1 and A[i] == 1:
            break
if cnt >= N:
    print(N)
else:
    print(cnt)
        

```

```python
N, M = map(int, input().split())
edges = [list(map(int, input().split())) for i in range(M)]
G = [list() for i in range(N + 1)] #引数揃えてある
for a, b in edges:
    G[b].append(a) #方向を逆転

import sys
sys.setrecursionlimit(120000)
def dfs(pos, G, visited):
    visited[pos] = True
    for i in G[pos]:
        if visited[i] == False:
            dfs(i, G, visited)

Q = int(input())
all_visited = [False] * (N + 1)
for _ in range(Q):
    s, v = map(int, input().split())
    if s == 1:
        new = v
        visited = [False] * (N + 1)
        dfs(new, G, visited)
        all_visited = [m or u for m, u in zip(all_visited, visited)]
    else:
        if all_visited[v] == False:
            print('No')
        elif all_visited[v] == True:
            print('Yes')
         
```
