```python
N = int(input())
grid = [[-1] * N for i in range(N)]
grid[0][(N - 1) // 2] = 1
from collections import deque
q = deque()
q.append([0, (N - 1) // 2, 1])
for i in range(N ** 2 - 1):
    r, c, k = q.popleft()
    r_ = (r - 1) % N 
    c_ = (c + 1) % N
    k_ = k + 1
    r__ = (r + 1) % N
    if grid[r_][c_] == -1:
        grid[r_][c_] = k_
        q.append([r_, c_, k_])
    else:
        grid[r__][c] = k_
        q.append([r__, c, k_])

for i in range(N):
    print(*grid[i])
    
```

```python
N, M = map(int, input().split())
from collections import defaultdict
d = defaultdict(lambda: defaultdict(int))
ans = 0
for j in range(M):
    r, c = map(int, input().split())
    if d[r - 1][c - 1] == d[r - 1][c] == d[r][c - 1] == d[r][c] == 0:
        d[r - 1][c - 1] = 1
        d[r - 1][c] = 1
        d[r][c - 1] = 1
        d[r][c] = 1
        ans += 1

print(ans)
```

```python
H, W = map(int, input().split())
grid = []
for _ in range(H):
    row = list(input())
    grid.append(row)
dist = []
for _ in range(H):
    row = [-1] * W
    dist.append(row)

from collections import deque, defaultdict

q = deque()
q.append([0, 0])
dist[0][0] = 0

portals = defaultdict(list)
for i in range(H):
    for j in range(W):
        if grid[i][j] != '.' and grid[i][j] != '#':
            portals[grid[i][j]].append([i, j])

while len(q) > 0:
    pos = q.popleft()
    y, x = pos[0], pos[1]
    next = [[y + 1, x],[y, x + 1], [y - 1, x],[y, x - 1]]
    if grid[y][x] in portals:
        for py, px in portals[grid[y][x]]:
            next.append([py, px])
        del portals[grid[y][x]]
    for ny, nx in next:
        if 0 <= ny < H and 0 <= nx < W:
            if grid[ny][nx] != '#' and dist[ny][nx] == -1:
                dist[ny][nx] = dist[y][x] + 1
                q.append([ny, nx])

print(dist[H - 1][W - 1])
  
```
