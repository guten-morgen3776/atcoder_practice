```python
N, M = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))

import bisect
mod = 998244353

B.sort()
rui_b = [0] * (M + 1)

for i in range(M):
    rui_b[i + 1] = (rui_b[i] + B[i]) % mod

ans = 0
for i in range(N):
    a = A[i]
    k = bisect.bisect_right(B, a)
    ans += (a * (2 * k - M) - rui_b[k] + (rui_b[M] - rui_b[k])) % mod


print(ans % mod)

```
