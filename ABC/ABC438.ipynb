```python
D, F = map(int, input().split())
i = 0
ans = 0
while F + 7 * i <= D:
    i += 1
    if F + 7 * i > D:
        ans = F + 7 * i - D
print(ans)
```

```python
N, M = map(int, input().split())
S = list(map(int, input()))
T = list(map(int, input()))

ans = float('inf')
for i in range(N - M + 1):
    dis = 0
    for j in range(M):
        dis += (S[i + j] - T[j] + 10) % 10
        #print(dis) 
    if dis < ans:
        ans = dis

print(ans)
```

```python
N = int(input())
A = list(map(int, input().split()))

deletes = 0
from collections import deque
stack = deque()

for i in range(N):
    if len(stack) > 0 and stack[-1][0] == A[i]:
        stack[-1][1] += 1
        if stack[-1][1] == 4:
            deletes += 1
            stack.pop()
    elif len(stack) == 0 or stack[-1][0] != A[i]:
        stack.append([A[i], 1])
    #print(stack)
print(len(A) - 4 * deletes)

    
```

```python
N = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
C = list(map(int, input().split()))

rui_A = [0] * (N + 1)
for i in range(N):
    rui_A[i + 1] = rui_A[i] + A[i]

rui_B = [0] * (N + 1)
for i in range(N):
    rui_B[i + 1] = rui_B[i] + B[i]

rui_C = [0] * (N + 1)
for i in range(N):
    rui_C[i + 1] = rui_C[i] + C[i]

max_f = -float('inf')
total_max = -float('inf')

for x in range(1, N - 1):
    f = rui_A[x] - rui_B[x]
    if f > max_f:
        max_f = f
    g = rui_B[x + 1] - rui_C[x + 1]
    if max_f + g > total_max:
        total_max = max_f + g

print(total_max + rui_C[N])
```
