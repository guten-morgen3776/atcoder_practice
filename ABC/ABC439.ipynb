```python
N = int(input())
from collections import defaultdict
cnt = defaultdict(int)

max_val = int(N ** 0.5)

for x in range(1, max_val):
    for y in range(x + 1, max_val + 1):
        val = x ** 2 + y ** 2
        if val <= N:
            cnt[val] += 1

ans = []
for n, c in cnt.items():
    if c == 1:
        ans.append(n)
print(len(ans))
ans.sort()
print(*ans)
```

```python
N = int(input())
A = list(map(int, input().split()))
from collections import Counter
cnt_l = Counter()
cnt_r = Counter(A)

ans = 0

for j in range(N):
    cnt_r[A[j]] -= 1
    if A[j] % 5 == 0:
        val_7 = A[j] // 5 * 7
        val_3 = A[j] // 5 * 3
        ans += cnt_r[val_7] * cnt_r[val_3]
        ans += cnt_l[val_7] * cnt_l[val_3]
    cnt_l[A[j]] += 1
print(ans)
```
