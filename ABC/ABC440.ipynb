```python
"""
リストcに対して周期2Wのうち前半のW個の要素の和を求める関数を作れば良い
これをT回繰り返すので愚直にやるとTLEする
周期2Wのうち前半W個の和を求めるのは累積和を使えばO(1)で求められる
周期2Wのうち前半W個の和を求める関数を作れば良い
でも累積和を計算するときにO(2W)かかるのでT回繰り返すとO(T*2W)でTLEする
なので累積和を1回だけ計算しておいて、各クエリではO(1)で答えを出すようにする 
でもリストCは各クエリで異なるので累積和を1回だけ計算するのは無理
なので各クエリで累積和を計算する必要がある
でも累積和を計算するのにO(2W)かかるのでT回繰り返すとO(T*2W)でTLEする
???
でも制約条件としてNの総和とWの総和が10^5以下なので、累積和を計算するのにかかる時間の総和も10^5以下になる
だからこのロジックに沿って実装すればいい
あとはxを選ぶことによって計算範囲をスライドさせることができるので、それぞれ計算して最小のものを求めればいい
"""
T = int(input())
for _ in range(T):
    N, W = map(int, input().split())
    C = list(map(int, input().split()))
    #2Wの長さに積み重ねる
    merged = [0] * (2 * W)
    for i in range(N):
        merged[i % (2 * W)] += C[i]
    # それぞれのxについて計算　円環なのでリスト二倍にしてスライドさせる
    merged = merged + merged
    #尺取り法で前半W個の和を求める
    cur_sum = sum(merged[:W])
    ans = cur_sum
    for x in range(2 * W):
        cur_sum = cur_sum - merged[x] + merged[x + W]
        if cur_sum < ans:
            ans = cur_sum     
    
    print(ans)
```
