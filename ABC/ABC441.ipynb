```python
#まず酒が入ってないやつを全部飲んで、そこから酒をXml以上飲めばいい
N, K, X = map(int, input().split())
A = list(map(int, input().split()))
#できるだけ小さいAiに酒が入っていれば良い Aをソートして下位K個に酒が入っているとする
#それから取っていって合計がXを超えたらおk
#計算量はo(K)
cur_amount = 0
cup = 0
A_ = sorted(A)
for i in range(K):
    cur_amount += A_[i]
    if cur_amount >= X:
        cup = i + 1
        break
if cup > 0:
    print(cup + N - K) #まず水から飲むからそれを足す
else:
    print(-1)
```

```python
#という上の考察は違くて、わからない中でできるだけ少ない回数にしたいのでできるだけ大きい容器を飲めばいい
#最初に水入りのN-K個を飲むので、残りのK個をできるだけ大きい順んい飲んでいく
#最悪の場合は酒に入っているAiができるだけ小さい場合なので、ソートして下位K個に酒が入っていて、
#それを大きい順に呑んで行ってXを超えればいい
#まず酒が入ってないやつを全部飲んで、そこから酒をXml以上飲めばいい

N, K, X = map(int, input().split())
A = list(map(int, input().split()))

cur_amount = 0
cup = 0
A_ = sorted(A)
for i in range(K):
    cur_amount += A_[K - i - 1]
    if cur_amount >= X:
        cup = i + 1
        break
if cup > 0:
    print(cup + N - K) #まず水から飲むからそれを足す
else:
    print(-1)
```

```python
N, M, L, S, T = map(int, input().split())

G = [[] for _ in range(N + 1)]
for _ in range(M):
    u, v, w = map(int, input().split())
    # タプル (行き先, 重み) として格納
    G[u].append((v, w))

#1~Nについて調べる
#同じとこ何回も通っていいから距離はどうでもいい　
#コストと回数を保持しながら動く
#全探索すれば4^10で大体10^6で通せる
#頂点１からスタートしてL回後でコストは条件満たす点が答え
#枝分かれをすべて表すにはdfs

import sys
sys.setrecursionlimit(10 ** 8)

ans = set()
def dfs(pos, cur_step, cur_cost):
    if cur_step == L:
        if S <= cur_cost <= T:
                ans.add(pos)
        return 
    else:
        for next, weight in G[pos]:
            dfs(next, cur_step + 1, cur_cost + weight)


dfs(1, 0, 0)
ans = sorted(list(ans))
print(*ans)
```
