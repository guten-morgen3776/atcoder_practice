```python
#iと利害関係がない人数が分かればいい
#毎回検索してたらo(N * M)になる
#利害関係をそのまま保持するのではなくて関係者人数のリストとして保持する

N, M = map(int, input().split())
from collections import defaultdict
num = defaultdict(int)
for _ in range(M):
    a, b = map(int, input().split())
    num[a] += 1
    num[b] += 1

ans = []
for i in range(N):
    n = N - num[i + 1] - 1 #本人を引くのを忘れないように
    if n >= 3:
        p = n * (n - 1) * (n - 2) // 6
    else:
        p = 0
    ans.append(p)
print(*ans)
```

```python
#区間和だから累積和
#でも入れ替えないといけないからあらかじめ計算したものをクエリ１の時に書き換える必要あり
#入れ替え後のAと累積和配列は保持し続けないといけない
#リスト要素の交換はA[i], A[j] = A[j], A[i]でできる
N, Q = map(int, input().split())
A = list(map(int, input().split()))
rui = [0] * (N + 1)
for i in range(1, N + 1):
    rui[i] = A[i - 1] + (rui[i - 1] if i > 1 else 0) 

for _ in range(Q):
    que = list(map(int, input().split()))
    if que[0] == 1:
        x = que[1]
        rui[x] -= A[x - 1] - A[x] 
        A[x - 1], A[x] = A[x], A[x - 1]
    else:
        l, r = que[1], que[2]
        ans = rui[r] - rui[l - 1]
        print(ans)

```
