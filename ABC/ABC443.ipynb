```python
"""
Aiの時には開いてない
つまり閉じている100秒間の間にしか通っていない、それがNかいある
青木くんが通る時に閉じればいいってこと？イエス
開く時間からそれ以上で最小のAiまでの時間の総和が求める時間
二分探索で求められるからo(NlogN)
"""
N, T = map(int, input().split())
A = list(map(int, input().split()))
time = 0
open = 0
import bisect
while open < T:
    #print(open)
    next_came_idx = bisect.bisect_right(A,open)
    if next_came_idx < N:
        next_came = A[next_came_idx]
    else:
        next_came = T
    #print(next_came)
    #print('=' * 10)
    time += next_came - open
    open = next_came + 100
    
print(time)
```

```python
"""
すべてのコマがの範囲内に入っていればいい
y方向は移動せずにx方向だけを考えればいい
隣との絶対値の差を１にしたい
一番xが小さいやつは動かさない
そこから左右方向に進んでいって差が１以内になるように動かしていく
駒の数はNこだからO(N)で行けるはず
でもこれだとmin_xが複数ある時におかしくなる
"""
T = int(input())
for _ in range(T):
    N = int(input())
    R = list(map(int, input().split()))
    #左から
    L = [0] * N
    L[0] = R[0]
    for i in range(1, N):
        L[i] = min(R[i], L[i - 1] + 1)
    #右から
    P = [0] * N
    P[N - 1] = R[N - 1]
    for j in range(N - 2, -1, -1):
        P[j] = min(R[j], P[j + 1] + 1)
    ans = 0
    for k in range(N):
        ans += R[k] -  min(L[k], P[k])
    print(ans)
```
