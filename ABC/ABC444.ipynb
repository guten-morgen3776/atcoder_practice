```python
"""
なんか累積和っぽい？
1+11+111+って感じだよね
多分全部数字に直して足しているとTLEだよね
先にながさNのリストを作ってそこに縦に足した時の和を保存
和はN...N-1...N-2ってなっていくはず
もし10を超えたら次のやつに一個足して10引くっていうのを繰り返す
それぞれの桁kにたいしてAiがk以上の個数を求める
Aについて度数分布を記録する配列を作って累積和すればこれは求められる
"""
N = int(input())
A = list(map(int, input().split()))
max_A = max(A)
A_ran = [0] * (max_A + 1)
for i in range(N):
    A_ran[A[i]] += 1
rui = [0] * (max_A + 1)
for j in range(1, max_A + 1):
    rui[max_A - j] = rui[max_A - j + 1] + A_ran[max_A - j + 1]
tate = [0] * max_A
for k in range(max_A):
    tate[k] = rui[k]

#繰り上がり
ans = []
carry = 0
for val in tate:
    current_val = val + carry
    ans.append(current_val % 10)
    carry = current_val // 10
while carry > 0:
    ans.append(carry % 10)
    carry //= 10
print(''.join(map(str, reversed(ans))))
```
