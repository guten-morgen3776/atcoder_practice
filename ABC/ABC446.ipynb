```python
"""
O(T + N)でいけるはず
D日ごに捨てるから仕入れた日を保持する必要がある
N日間は挙動を追う必要がある
キューを使う
"""
from collections import deque
T = int(input())
for _ in range(T):
    N, D = map(int, input().split())
    A = list(map(int, input().split()))
    B = list(map(int, input().split()))
    q = deque()
    for i in range(N):
        q.append([A[i], i])
        cons = B[i]
        while cons > 0:
            cnt, day = q[0]
            if cons <= cnt:
                q[0][0] -= cons
                cons = 0
            else:
                cons -= cnt
                q.popleft()
        while q and q[0][1] <= i - D:
            q.popleft()
    ans = 0
    for j in range(len(q)):
        ans += q[j][0]
    print(ans)


        





    
```

```python
"""
Aの中でA[i] + 1 = A[i + 1]を満たす部分を探せばいい、コレだけ？
違うわ、連続している必要ない、繋ぎ合わせてもいいんだ、Aで隣り合っているかどうかは関係ない
始点が決まればその部分列の長さは求められる
すべての始点で試すとO(N*N)
事前になんかしらの配列計算が必要なはず
あり得るオーダーはO(N),O(NlogN)のどっちか
入れ替えたらダメだからソートはないよね
だとしたらO(N)
"""
from collections import defaultdict
N = int(input())
A = list(map(int, input().split()))
dp = defaultdict(int)
for a in A:
    dp[a] = dp[a - 1] + 1
print(max(dp.values()))
```
