```python
N, M = map(int, input().split())
grid_1 = [list(input()) for _ in range(N)]
grid_2 = [list(input()) for _ in range(M)]

for i in range(N - M + 1):
    for j in range(N - M + 1):
        mini_grid = [row[j : j + M] for row in grid_1[i : i + M]]
        if grid_2 == mini_grid:
            print(i + 1, j + 1)
```
