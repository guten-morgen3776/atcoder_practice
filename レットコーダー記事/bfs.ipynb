```python
from collections import deque
R, C = map(int, input().split())
sy, sx = map(int, input().split())
gy, gx = map(int, input().split())
grid = []
for _ in range(R):
    row = list(input())
    grid.append(row)
q = deque()
dist = [[-1] * C for _ in range(R)] #距離配列　最初は全部ー１

sx = sx - 1
sy = sy - 1
gx = gx - 1
gy = gy - 1

q.append((sy, sx))
dist[sy][sx] = 0

dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

while q: #qに中身がある間は繰り返す
    y, x = q.popleft()
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if not (0 <= nx < C and 0 <= ny < R):
            continue
        if grid[ny][nx] == '#':
            continue
        if dist[ny][nx] != -1:
            continue
        dist[ny][nx] = dist[y][x] + 1
        q.append((ny, nx))
print(dist[gy][gx])
```
