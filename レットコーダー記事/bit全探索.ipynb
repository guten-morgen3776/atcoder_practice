```python
N, M = map(int, input().split())

swiches = []
for _ in range(M):
    row = list(map(int, input().split()))
    swiches.append(row)
P = list(map(int, input().split()))

ans = 0
for i in range(1 << N):
    cur_subset = []
    for j in range(N):
        if (i &(1 << j)):
            cur_subset.append(j) #オンになってるスイッチ

    light_cnt = 0
    for k in range(M):
        on_swiches = 0
        for l in range(1, swiches[k][0] + 1):
            if (swiches[k][l] - 1) in cur_subset:
                on_swiches += 1
        if on_swiches % 2 == P[k]:
            light_cnt += 1
    if light_cnt == M:
        ans += 1
print(ans)          

```
