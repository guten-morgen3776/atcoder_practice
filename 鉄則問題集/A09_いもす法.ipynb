```python
H, W, N = map(int, input().split())
grid = [[0] * (W + 2) for i in range(H + 2)] #一個余裕持たせる まじで大事！！！
for _ in range(N):
    a, b, c, d = map(int, input().split())
    grid[a][b] += 1
    grid[a][d + 1] -= 1
    grid[c + 1][b] -= 1
    grid[c + 1][d + 1] += 1

#横方向
for i in range(1, H + 1):
    for j in range(1, W + 1):
        grid[i][j] = grid[i][j] + grid[i][j - 1]
#縦方向
for j in range(1, W + 1):
    for i in range(1, H + 1):
        grid[i][j] = grid[i][j] + grid[i - 1][j]

for i in range(1, H + 1):
    print(*grid[i][1 : H+1])

    
```
