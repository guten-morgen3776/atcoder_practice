```python
N, X = map(int, input().split())
A = list(map(int, input().split()))


def binary_search(X, A):
    left = 0
    right = N - 1
    mids = []   
    while left <= right:
        mid = (right + left) // 2
        mids.append(mid)
        if X < A[mid]:
            right = mid - 1
        if X > A[mid]:
            left = mid + 1
        if X == A[mid]:
            return mid
    return -1

ans = binary_search(X, A)
print(ans + 1)
```
