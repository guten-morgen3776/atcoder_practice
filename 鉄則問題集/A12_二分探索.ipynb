```python
N, K = map(int, input().split())
A = list(map(int, input().split()))

def check(x, N, K, A): #x秒後
    sum = 0
    for i in range(N):
        sum += x // A[i] #プリンターiが出力する枚数
    if sum >= K:
        return True
    return False

left = 1
right = 10 ** 9
while left < right:
    mid = (left + right) // 2
    ans = check(mid, N, K, A)

    if ans == False:
        left = mid + 1
    if ans == True:
        right = mid

print(left, right)
```
