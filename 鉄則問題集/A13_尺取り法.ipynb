```python
N, K = map(int, input().split())
A = list(map(int, input().split()))

right = 1
ans = 0

for left in range(N): #左を動かす
    while right < N and A[right] - A[left] <= K: #右を動かす
        right += 1
    #print(f'left: {left}, right: {right}')
    ans += right - left - 1
    
print(ans)   
```

```python
#二分探索でとく
N, K = map(int, input().split())
A = list(map(int, input().split()))
A.sort()
import bisect

ans = 0
for i in range(N):
    ans += bisect.bisect_right(A, i + K)
    #print(ans)
print(ans - 1)
```
