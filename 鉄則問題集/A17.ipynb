```python
N = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
dp = [0] * (N + 1)
dp[1] = 0
dp[2] = A[0]
for i in range(3, N + 1):
    dp[i] = min(dp[i - 2] + B[i - 3], dp[i - 1] + A[i - 2])

answer = []
pos = N
while True:
    answer.append(pos)
    if pos == 1:
        break
    if dp[pos] == dp[pos - 1] + A[pos - 2]:
        pos =  pos - 1
    else:
        pos = pos - 2
answer.reverse()
print(len(answer))
print(*answer)
```
