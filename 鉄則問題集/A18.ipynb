```python
N, S = map(int, input().split())
A = list(map(int, input().split()))
dp = []
for _ in range(N + 1):
    row = [0] * (S + 1)
    dp.append(row)
dp[0][0] = 1
for i in range(1, N + 1):
    Ai = A[i - 1]
    for j in range(S + 1):
        if j - Ai >= 0:
            dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - Ai])
        else:
            dp[i][j] = dp[i - 1][j]

if dp[N][S] == 1:
    print('Yes')
else:
    print('No')

```
