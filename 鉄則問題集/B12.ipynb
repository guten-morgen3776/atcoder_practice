```python
N = int(input())

def check(N, x):
    if x ** 3 + x < N:
        return False
    return True

left = 1
right = 100000

while right > left:
    mid = (right + left) // 2
    ans = check(N, mid)
    if ans == True:
        right = mid
    if ans == False:
        left = mid + 1
    
print(left)

    
```
