```python
import sys
sys.setrecursionlimit(120000)

N, M = map(int, input().split())
edges = [list(map(int, input().split())) for i in range(M)]
G = [list() for i in range(N + 1)]
for a, b in edges:
    G[a].append(b)
    G[b].append(a)

visited = [False] * (N + 1)
ans = []

def dfs(pos, path):
    global ans
    if ans:
        return 
    visited[pos] = True
    path.append(pos)
    if pos == N:
        ans = list(path)
        return
    for i in G[pos]:
        if not visited[i]:
            dfs(i, path)
    path.pop() #バックトラック

dfs(1, [])

print(*ans)
```
