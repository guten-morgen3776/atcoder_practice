```python
R, C = map(int, input().split())
sy, sx = map(int, input().split())
gy, gx = map(int, input().split())

grid = []
for _ in range(R):
    row = list(input())
    grid.append(row)

dist = []
for _ in range(R):
    row = [-1] * C
    dist.append(row)

import numpy as np
grid = np.array(grid)
dist = np.array(dist)

from collections import deque
q = deque()

sy = sy - 1
sx = sx - 1
gy = gy - 1
gx = gx - 1

dist[sy, sx] = 0
q.append([sy, sx])

while len(q) > 0:
    pos = q.popleft()
    y, x = pos[0], pos[1]
    next = [[y + 1, x],[y, x + 1], [y - 1, x],[y, x - 1]]
    for i in next:
        yi = i[0]
        xi = i[1]
        if dist[yi, xi] == -1 and grid[yi, xi] == '.':
            dist[yi, xi] = dist[y, x] + 1
            q.append(i)

print(dist[gy, gx])

    
```
